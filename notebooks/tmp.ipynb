```python
import numpy as np
from matplotlib import pyplot as plt

import xrt.backends.raycing.materials as rm
from IPython.display import display, HTML
```

```python
mSi = rm.Material('Si', rho=2.33, kind='lens')
mNi = rm.Material('Ni', rho=8.9, kind='lens')
en = 12.4e3  # eV
# abs_len = 10. / mSi.get_absorption_coefficient(en) # mm
th = 4.  # mcm
# example from the paper
print(
    np.exp(-mNi.get_absorption_coefficient(en) * th * 1e-4)
)
print(
    -np.log(0.6585359830135198) / (mNi.get_absorption_coefficient(en) * 1e-4)
)

ens = np.linspace(5e3, 20e3)  # eV
absorps = [0.2, 0.4, 0.6]
plt.figure()
for absorp in absorps:
    plt.plot(ens*1e-3, -np.log(1 - absorp) / (mSi.get_absorption_coefficient(ens) * 1e-4), label=("Поглощение %d%%" % int(100 * absorp)))
plt.legend()
plt.ylabel('Толщина образца, [мкм]')
plt.xlabel('Энергия фотонов, [кэВ]')
plt.show()
```
